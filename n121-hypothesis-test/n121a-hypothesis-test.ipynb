```python
import pandas as pd
import numpy as np
url = 'https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/seoul_tree/seoul_tree.txt'
dataset = pd.read_csv(url, sep = "\t")
# í—¤ë”í–‰ ì‚­ì œ
dataset.columns = dataset.iloc[0]
dataset.columns.name=None
dataset.drop(0, axis=0, inplace=True)
dataset.drop(['ê¸°ê°„'], axis=1, inplace=True)
dataset
```

### "ì„œìš¸ì‹œì—ëŠ” êµ¬ë³„ë¡œ í‰ê·  ì•½ 400 ê·¸ë£¨ ì •ë„ì˜ `ì´íŒë‚˜ë¬´`ê°€ ìˆë‹¤"ë¼ëŠ” ê°€ì„¤ì— ëŒ€í•´ ê°€ì„¤ê²€ì •ì„ ì‹œí–‰í•˜ì„¸ìš”.

- ë‹¨, `ì„œìš¸ì‹œ`, `ê³µì›ë…¹ì§€ì‚¬ì—…ì†Œ`, `ì‹œì„¤ê´€ë¦¬ê³µë‹¨` ë¶€ë¶„ì˜ ë°ì´í„°ëŠ” ì œì™¸í•©ë‹ˆë‹¤.
- ê°€ì„¤ê²€ì • ê²°ê³¼ì˜ pvalue ê°’ì„ `pv1`ì— ì €ì¥í•˜ì„¸ìš”. (`.pvalue` ì‚¬ìš©)

```python
from scipy import stats
from scipy.stats import ttest_ind
dataset2 = dataset # ì„œìš¸ì‹œ, ê³µì›ë…¹ì§€ì‚¬ì—…ì†Œ, ì‹œì„¤ê´€ë¦¬ê³µë‹¨ë¥¼ ì œê±°í•œ ë°ì´í„° ìƒì„±
dataset2.drop(index=1, inplace=True)
dataset2.drop(index=27, inplace=True)
dataset2.drop(index=28, inplace=True)
def toInt(string): # str -> int í•¨ìˆ˜
  return int(string.replace(',',''))
dataset2 = dataset2.set_index('ìì¹˜êµ¬') # ìì¹˜êµ¬ë¥¼ ì¸ë±ìŠ¤ë¡œ ë§Œë“¤ê¸°
dataset2
dataset2['ì´íŒë‚˜ë¬´'] = dataset2['ì´íŒë‚˜ë¬´'].apply(toInt) # ì´íŒë‚˜ë¬´ object -> int

d1 = dataset2['ì´íŒë‚˜ë¬´']
d2 = 400
pv1 = stats.ttest_1samp(dataset2['ì´íŒë‚˜ë¬´'], 400)
pv1
```

### ì´ë²ˆì—ëŠ” 10ê°œì˜ êµ¬ë¥¼ ìƒ˜í”Œë§í•˜ì—¬, ìœ„ì— ëŒ€í•œ ê°€ì„¤ê²€ì •ì„ ë‹¤ì‹œ í•œë²ˆ ì‹œí–‰í•´ë³´ì„¸ìš”.

- ê°€ì„¤ê²€ì • ê²°ê³¼ì˜ pvalueê°’ì„ `pv2`ì— ì €ì¥í•˜ì„¸ìš”


```python
import numpy as np
np.random.seed(123)

dataset2_sp = np.random.choice(dataset2["ì´íŒë‚˜ë¬´"], size = 10)  #dataset2ì˜ ì´íŒë‚˜ë¬´ê°’ 10ê°œë¥¼ ëœë¤ ì¶”ì¶œ
pv2 = stats.ttest_1samp(dataset2_sp, 400)
pv2
```
