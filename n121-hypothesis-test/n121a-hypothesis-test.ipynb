```python
import pandas as pd
import numpy as np
url = 'https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/seoul_tree/seoul_tree.txt'
dataset = pd.read_csv(url, sep = "\t")
# Ìó§ÎçîÌñâ ÏÇ≠Ï†ú
dataset.columns = dataset.iloc[0]
dataset.columns.name=None
dataset.drop(0, axis=0, inplace=True)
dataset.drop(['Í∏∞Í∞Ñ'], axis=1, inplace=True)
dataset
```

### "ÏÑúÏö∏ÏãúÏóêÎäî Íµ¨Î≥ÑÎ°ú ÌèâÍ∑† ÏïΩ 400 Í∑∏Î£® Ï†ïÎèÑÏùò `Ïù¥ÌåùÎÇòÎ¨¥`Í∞Ä ÏûàÎã§"ÎùºÎäî Í∞ÄÏÑ§Ïóê ÎåÄÌï¥ Í∞ÄÏÑ§Í≤ÄÏ†ïÏùÑ ÏãúÌñâÌïòÏÑ∏Ïöî.

- Îã®, `ÏÑúÏö∏Ïãú`, `Í≥µÏõêÎÖπÏßÄÏÇ¨ÏóÖÏÜå`, `ÏãúÏÑ§Í¥ÄÎ¶¨Í≥µÎã®` Î∂ÄÎ∂ÑÏùò Îç∞Ïù¥ÌÑ∞Îäî Ï†úÏô∏Ìï©ÎãàÎã§.
- Í∞ÄÏÑ§Í≤ÄÏ†ï Í≤∞Í≥ºÏùò pvalue Í∞íÏùÑ `pv1`Ïóê Ï†ÄÏû•ÌïòÏÑ∏Ïöî. (`.pvalue` ÏÇ¨Ïö©)

```python
from scipy import stats
from scipy.stats import ttest_ind
dataset2 = dataset # ÏÑúÏö∏Ïãú, Í≥µÏõêÎÖπÏßÄÏÇ¨ÏóÖÏÜå, ÏãúÏÑ§Í¥ÄÎ¶¨Í≥µÎã®Î•º Ï†úÍ±∞Ìïú Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
dataset2.drop(index=1, inplace=True)
dataset2.drop(index=27, inplace=True)
dataset2.drop(index=28, inplace=True)
def toInt(string): # str -> int Ìï®Ïàò
  return int(string.replace(',',''))
dataset2 = dataset2.set_index('ÏûêÏπòÍµ¨') # ÏûêÏπòÍµ¨Î•º Ïù∏Îç±Ïä§Î°ú ÎßåÎì§Í∏∞
dataset2
dataset2['Ïù¥ÌåùÎÇòÎ¨¥'] = dataset2['Ïù¥ÌåùÎÇòÎ¨¥'].apply(toInt) # Ïù¥ÌåùÎÇòÎ¨¥ object -> int

d1 = dataset2['Ïù¥ÌåùÎÇòÎ¨¥']
d2 = 400
pv1 = stats.ttest_1samp(dataset2['Ïù¥ÌåùÎÇòÎ¨¥'], 400)
pv1
```

### Ïù¥Î≤àÏóêÎäî 10Í∞úÏùò Íµ¨Î•º ÏÉòÌîåÎßÅÌïòÏó¨, ÏúÑÏóê ÎåÄÌïú Í∞ÄÏÑ§Í≤ÄÏ†ïÏùÑ Îã§Ïãú ÌïúÎ≤à ÏãúÌñâÌï¥Î≥¥ÏÑ∏Ïöî.

- Îã®, ÏïÑÎûòÏùò ÏΩîÎìúÎ•º Ï∞∏Ï°∞ÌïòÏó¨ sampleÏùÑ ÏúÑÌïú seed Îäî 123Î°ú ÏÑ§Ï†ïÌïòÏÑ∏Ïöî.
- Í∞ÄÏÑ§Í≤ÄÏ†ï Í≤∞Í≥ºÏùò pvalueÍ∞íÏùÑ `pv2`Ïóê Ï†ÄÏû•ÌïòÏÑ∏Ïöî


```python
import numpy as np
np.random.seed(123)

dataset2_sp = np.random.choice(dataset2["Ïù¥ÌåùÎÇòÎ¨¥"], size = 10)  #dataset2Ïùò Ïù¥ÌåùÎÇòÎ¨¥Í∞í 10Í∞úÎ•º ÎûúÎç§ Ï∂îÏ∂ú
pv2 = stats.ttest_1samp(dataset2_sp, 400)
pv2
```
